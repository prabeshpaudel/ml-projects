```python
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(n_samples=1000, noise=0.05)
```

```python
dbscan = DBSCAN(eps=0.05, min_samples=5)
```

```python
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
dbscan.core_sample_indices_
```
